# **WebScraping_and_wordcloud_news**


```python
#import libs
import pandas as pd
import feedparser as ps
import nltk 
import wordcloud 
import numpy as np
```

```python
feeds=['https://www.infomoney.com.br/feed/','https://investnews.com.br/feed/','https://www.jornaldenegocios.pt/rss','https://bmcnews.com.br/feed/','https://exame.com/invest/feed/']
x=feeds[4]
news_title_x=[]
for x in feeds:
   print(x)
   _x=ps.parse(x)
   #Selecionando Title _x
     
   for post in _x.entries:
      news_title_x.append(post['title'])
      news_title_x.append(post['description'])
news_title_x
```

```python
#Separando em frases
from nltk import sent_tokenize
text_x=''

token_x2=[]
token_x3=[]

#tratando token e retirando urls
for i in range(len(news_title_x)):
  if 'href' not in news_title_x[i]:
    token_x2.append(news_title_x[i])
  else:
     token_x3.append(news_title_x[i])
print(token_x2)
print(len(token_x2))
print(token_x3)
print(len(token_x3))
```

```python
#limpando token_x3
token_x3_trat=[]

for j in range(len(token_x3)):#Percorre cada frase
    #print('frase {}'.format(j))    
    d=[]
    for i in range(len(token_x3[j])): #procura os caracteres < e > e guarda na lista d que é uma lista de bads
        if token_x3[j][i]=="<":
            for i2 in range(len(token_x3[j])-(i+1)):
                if token_x3[j][i2]==">":
                    d.append(i)
                    d.append(i2)
        d=list(set(d))
        d.sort()
        d.append(len(token_x3[j]))
    
    c=[] #esta lista terá os indices das letras filtradas
    if d[0]>0: #se d não começar em zero adiciona (na lista C) um range de 0 até primeiro bad de d 
        c.append(np.arange(0,d[0]))
    for k in range(len(d)//2-1): #vai colocar em C listas de numeros que ficam entre os intervalos de bads marcados em d
        
        c.append(np.arange(d[2*k+1]+1,d[2*k+2]-1))  
    
    token_x3_ext=''
    #print('c:',c)
    for i in range(len(c)): #concatenar as listas de C
        #print('tamanoci',len(c[i]))
        for l in range(len(c[i])):
            token_x3_ext=token_x3_ext+token_x3[j][c[i][l]]
    #print(token_x3_ext)
    token_x3_trat.append(token_x3_ext)           

token_x3_trat
```

```python
#Concatenando os Tokens
token_concat=[]
token_concat.append(token_x2)
token_concat.append(token_x3_trat)
token_concat
```

```python
#colocando os tokens em uma lista de uma dimensão
tokens=[]
for j in token_concat:
    for i in j:
        tokens.append(i)
tokens
```

```python
df_tokens=pd.DataFrame(tokens,columns=['token'])
df_tokens.to_csv('tokens_22_08_2023.csv',index=False)
```
