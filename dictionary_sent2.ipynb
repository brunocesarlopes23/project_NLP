```python
import pandas as pd
```

```python
sentilex=pd.read_csv('sentiLex_lem_PT02.csv')
sentilex
```

```python
oplexicon2=pd.read_csv('oplexicon_v2.1.csv')
oplexicon2
```

```python
oplexicon3=pd.read_csv('oplexicon_v3.0.csv',encoding='cp850')
oplexicon3
```

```python
a=set(sentilex['term'].values)
b=set(oplexicon3['term'].values)
c=set(a).intersection(b);
```

```python
indice=[]
for i in c:
    indice.append(sentilex.loc[sentilex['term']==i].index[0])
```

```python
sentilex2=sentilex[['term','grammar_category','polarity','polarity_classification']].drop(labels=indice,axis=0)
sentilex2.rename(columns={'grammar_category':'type','polarity_classification':'polarity_revision'},inplace=True)
df_dict=pd.concat([oplexicon3,sentilex2],ignore_index=True)
df_dict
```

```python
df_dict.to_csv('dictionary_lex_PT.csv',index=False)
```
